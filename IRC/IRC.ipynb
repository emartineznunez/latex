```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from math import sqrt
from scipy.stats import linregress
from scipy.interpolate import make_interp_spline


#reading the conductivities from a file
irc = pd.read_csv("IRC.csv",sep=',')
rxn = np.array(irc['rxn'].values)
energy = np.array(irc['energy'].values)

plt.rcParams['text.usetex'] = True

plt.plot(rxn,energy, '-',color='black')

plt.xlabel(r'Reaction coordinate ($\sqrt{amu}\times a_{0}$)',fontsize=18)
plt.ylabel('Energy (kcal/mol)',fontsize=18)

#plt.xlim([1,30])
plt.ylim([-120,19])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
#plt.title('GLN',x=0.8, y=0.8, fontsize=16)
plt.savefig('IRC.svg')
plt.show()


plt.show()
```
