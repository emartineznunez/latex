```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from math import sqrt
from scipy.stats import linregress
from scipy.interpolate import make_interp_spline


#reading the conductivities from a file
citations = pd.read_csv("BR1.csv",sep=',')
h2 = np.array(citations['H2+C2HNO'].values)
chn = citations['CHN+CH2O'].values
h2o = citations['H2O+C2HN'].values
x = np.array(citations['iter'].values)
x_ = np.linspace(x.min(), x.max(), 500)

h2_interp = make_interp_spline(x,h2)
h2_i = h2_interp(x_)
chn_interp = make_interp_spline(x,chn)
chn_i = chn_interp(x_)
h2o_interp = make_interp_spline(x,h2o)
h2o_i = h2o_interp(x_)


citations2 = pd.read_csv("BR2.csv",sep=',')
h2r = citations2['H2+C2H2NO'].values
chnr = citations2['CHN+CH3O'].values
hr = citations2['H+C2H3NO'].values
h2or = citations2['H2O+C2H2N'].values

h2r_interp = make_interp_spline(x,h2r)
h2r_i = h2r_interp(x_)
chnr_interp = make_interp_spline(x,chnr)
chnr_i = chnr_interp(x_)
h2or_interp = make_interp_spline(x,h2or)
h2or_i = h2or_interp(x_)
hr_interp = make_interp_spline(x,hr)
hr_i = hr_interp(x_)

plt.rcParams['text.usetex'] = True

plt.plot(x_,h2_i, '-',color='black',label='H$_{2}$ + C$_{2}$HNO')
plt.plot(x_,chn_i, '-',color='red',label='HCN + CH$_{2}$O')
plt.plot(x_,h2o_i, '-',color='blue',label='H$_{2}$O + C$_{2}$HN')



plt.xlabel('Iteration',fontsize=18)
plt.ylabel('Population',fontsize=18)

plt.legend(loc=2,frameon=False)
plt.xlim([1,30])
plt.ylim([-5,100])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('GLN',x=0.9, y=0.9, fontsize=20)
plt.savefig('BR1.svg')
plt.show()


plt.plot(x_,h2r_i, '-',color='black',label='H$_{2}$ + C$_{2}$H$_{2}$NO')
plt.plot(x_,chnr_i, '-',color='red',label='HCN + CH$_{3}$O')
plt.plot(x_,h2or_i, '-',color='blue',label='H$_{2}$O + C$_{2}$H$_{2}$N')
plt.plot(x_,hr_i, '-',color='green',label='H + C$_{2}$H$_{3}$NO')

plt.xlabel(r'Iteration',fontsize=18)
plt.ylabel(r'Population',fontsize=18)

plt.legend(loc=2,frameon=False)
plt.xlim([1,30])
plt.ylim([-5,100])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('GLN-H',x=0.9, y=0.9, fontsize=20)
plt.savefig('BR2.svg')
plt.show()
```
